<h3>In the below section the data distribution is positive skewed. so in order to get the normal distribution curve `log1p` function is applied</h3>

```python
df['count_log']=np.log1p(df['count'])
df['count_log']
```

```python
plt.hist(df['count_log'], bins=50, label='count_log')
```

```python
df['season_sin']=np.sin((2 * np.pi * df['season']) / 4)
df['season_cos']=np.cos((2 * np.pi * df['season']) / 4)
df.head()
```

```python
df['season'].plot()
```

<h3>Here the above graph concludes that the distribution of the data of seasons are increasing and decreasing. which is bad for the machine learning model because seasons should be treated equally as a cyclic motion not as some random increasing and decreasing values.</h3>

```python
df.plot.scatter('season_sin', 'season_cos').set_aspect("equal")
```

<h3> Here finally sin and cos function is applied on the season column so that while plotting it can show the cyclic view of the seasons</h3>

```python
df['new_datetime']=pd.to_datetime(df['datetime'], format="%Y-%m-%d %H:%M:%S")
df.head()
```

```python
df["year"] = df["new_datetime"].dt.year
df["month"] = df["new_datetime"].dt.month
df["day"] = df["new_datetime"].dt.day
df["hour"] = df["new_datetime"].dt.hour
df.head()
```

<h3>Repeating the same steps for column: month</h3>

```python
df['month'].unique().shape
```

```python
df['month_sin']=np.sin((2 * np.pi * df['month'])/12)
df['month_cos']=np.cos((2 * np.pi * df['month'])/12)
```

```python
df.head()
```

```python
df.plot.scatter('month_sin','month_cos').set_aspect('equal')
```

<h3>Repeating the same steps for column: hour</h3>

```python
df['hour'].unique().shape
```

```python
df['hour_sin']=np.sin((2 * np.pi * df['hour'])/24)
df['hour_cos']=np.cos((2 * np.pi * df['hour'])/24)
df.head()
```

```python
df.plot.scatter('hour_sin','hour_cos').set_aspect('equal')
```

<h3>Repeating the same steps for column: day</h3>

```python
df['day'].unique().shape
```

```python
df['day_sin']=np.sin((2 * np.pi * df['hour'])/19)
df['day_cos']=np.cos((2 * np.pi * df['hour'])/19)
df.head()
```

```python
df.plot.scatter('day_sin','day_cos').set_aspect('equal')
```
