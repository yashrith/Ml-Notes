```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x=np.linspace(0,1,num=500)
y=np.linspace(0,1,num=500)
```

```python
X,Y=np.meshgrid(x,y)
Z = X + Y
```

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(projection='3d')
s = ax.plot_surface(X, Y, Z, cmap=plt.cm.plasma)
cset=ax.contour(X,Y,Z, zdir='z', offset=0, cmap=plt.cm.plasma)
fig.colorbar(s, shrink=0.3)
```

```python
x=np.arange(-10,10,0.5)
y=np.arange(-10,10,0.5)
X,Y= np.meshgrid(x,y)
Z = X**2 + Y**2
```

<p><b>use the gradient descent and go down to find the minimum</b></p>

```python
fig=plt.figure(figsize=(8,8))
ax=fig.add_subplot(projection='3d')
s = ax.plot_surface(X, Y, Z, cmap=plt.cm.inferno)
cset=ax.contour(X,Y,Z, zdir='z', offset=0, cmap=plt.cm.inferno)
fig.colorbar(s, shrink=0.3)
```
