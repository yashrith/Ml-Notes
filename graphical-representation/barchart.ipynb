```python
mu=100 ### mean = 100
sigma = 15 ## std = 15
x=mu + sigma * np.random.randn(10000)
```

```python
num_bins = 20
plt.hist(x, bins=num_bins, edgecolor='white', facecolor='red')
```

```python
from scipy.stats import norm
```

<p>here the norm.pdf indicates the red line</p>

```python
n, bins, patches = plt.hist(x,bins=num_bins, edgecolor='white', facecolor='green', density=True)
y=norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
plt.xlabel('smarts')
plt.ylabel('probability')
plt.title('Histogram for smart students')
plt.show()
```
