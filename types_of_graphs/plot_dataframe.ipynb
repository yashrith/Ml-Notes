```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=pd.DataFrame(3*np.random.rand(50,4), columns=['a','b','c','d'])
df.head()
```

```python
df.plot.scatter(x='a',y='b', color='yellow', label='Group 1')
df.plot.scatter(x='c',y='d', color='white', label='Group 2')
```

```python
df=pd.DataFrame(3*np.random.rand(10,4), columns=['a','b','c','d'])
df.head()
```

```python
result=df.plot.scatter(x='a',y='b', color='yellow', label='Group 1')
df.plot.scatter(x='c',y='d', color='white', label='Group 2',ax=result)
```

```python
df.plot.bar()
```

```python
df.plot.bar(stacked=True)
```

```python
df.plot.barh(stacked=True)
```

```python
df.plot.box()
```

```python
df.plot.box(color={'boxes':'Cyan', 'whiskers':'Magenta', 'medians':'Yellow', 'caps':'White'})
```
