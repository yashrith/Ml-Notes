```python
import numpy as np
import matplotlib.pyplot as plt
```

# y = x^2

```python
x1=np.array([-3,-2,-1,0,1,2,3])
x2=np.array([9,4,1,0,1,4,9])
```

```python
plt.scatter(x1,x2,c='w')
plt.xlabel('x1')
plt.ylabel('x2')
```

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
```

```python
plt.plot(x,y, 'r--p', label='y=2x')
plt.legend()
plt.title('This is a demo from plot')
plt.xlabel('X')
plt.ylabel('Y')
```

```python
plt.plot(x1,x2, 'b:D', label='y=x^2')
plt.legend()
plt.title('This is a demo from plot')
plt.xlabel('X1')
plt.ylabel('X2')
```

<p>here plt.legend indicates the label to top right corner of the graph</p>

```python
plt.subplot(211)
plt.plot(x,y, 'r--p', label='y=2x')
plt.legend()

plt.subplot(212)
plt.plot(x1,x2, 'b:D', label='y=x^2')
plt.legend()
plt.savefig('image.png')
```
